```python
import random
# Initialize inventory and money 🎒💰
inventory = {}
total_money = 0

# Fish types and prices 🐟
fish_prices = {
   "Rohu": 12,
    "Catla": 18,
    "Sinchu": 25
}
print("\n🎣 --- Fishing Game Menu --- 🎣")
print("1️⃣ Catch Fish")
print("2️⃣ View Inventory")
print("3️⃣ Sell Fish")
print("4️⃣ Exit 🚪")

# Game loop 🔁
while True:
    choice = int(input("Choose an option: "))

    # Catch Fish 🐠
    if choice == 1:
        fish = random.choice(list(fish_prices.keys()))
        inventory[fish] = inventory.get(fish, 0) + 1
        print(f"🎉 You caught a {fish}!")

    # View Inventory 🎒
    elif choice == 2:
        if inventory == {}:       #if len(inventory) == 0:      #if not inventory:
            print("❌ No fish.")
        else:
            print("📦 Your Inventory:")
            for fish, count in inventory.items():
                print(f"{fish}: {count}")
# Sell Fish 💵
    elif choice == 3:
        if inventory == {}:
            print("❌ No fish to sell.")
        else:
            earnings = 0
            for fish, count in inventory.items():
                earnings += fish_prices[fish] * count
            total_money += earnings  # total_money = total_money earnings
            inventory.clear()
            print(f"💰 You sold all fish for ${earnings}!")

    # Exit 🚪
    elif choice == 4:
        print(f"🏁 Final money earned: ${total_money}")
        print("👋 Thanks for playing!")
        break

    else:
        print("⚠️ Invalid choice. Try again.")
```

```python
#logistic regresion
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset=pd.read_csv('Product_purchase.csv')

# Defining features (X) and target (Y)
X = dataset.iloc[:, [2, 3]].values # Taking 'Age' and 'EstimatedSalary' as features
Y = dataset.iloc[:, 4].values    # Taking 'Purchased' as the target

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression # Import LogisticRegression

# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)

# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Training the Logistic Regression model on the Training set
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Now, y_test and y_pred are defined, so classification_report can be used
print("report",classification_report(y_test,y_pred))
```

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train # Corrected x_train_std to X_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                color = ListedColormap(('red', 'green'))(i), label = j) # Changed 'c' to 'color'
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
#random forest
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#importing the datasets
datasets=pd.read_csv('Social_Network_Ads.csv')
X= datasets.iloc[:,[2,3]].values
Y= datasets.iloc[:,4].values
```

```python
#checking the column names
print(datasets.columns[[2,3]])
print(datasets.columns[4])
```

```python
#spliting the datasets into training set and test set
from sklearn.model_selection import train_test_split
X_Train,X_Test,Y_Train,Y_Test=train_test_split(X,Y,test_size=0.25,random_state=0)


#feature scaling
from sklearn.preprocessing import StandardScaler
sc_X= StandardScaler()
X_Train=sc_X.fit_transform(X_Train)
X_Test=sc_X.transform(X_Test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier(criterion='entropy',max_depth=None,
                              min_samples_split=5, min_samples_leaf=2)

#training the model
model.fit(X_Train,Y_Train)
print(model)

#predict the testing model
y_pred=model.predict(X_Test)
print(y_pred)

#creating confudion mattrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_Test,y_pred)
print(cm)

from sklearn.metrics import accuracy_score
print("Decision tree", accuracy_score(Y_Test,y_pred))
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
modellr= LogisticRegression(random_state=0)

#training the model
modellr.fit(X_Train,Y_Train)
print(modellr)

#predict the testing model
y_pred_lr=modellr.predict(X_Test)
print(y_pred_lr)

#creating confudion mattrix
from sklearn.metrics import confusion_matrix
cmlr=confusion_matrix(Y_Test,y_pred_lr)
print(cmlr)

from sklearn.metrics import accuracy_score
print("Logistic regression", accuracy_score(Y_Test,y_pred_lr))
```
